### TensorFlow Constants

    A TensorFlow constant is a tensor whose values cannot be changed after creation. It is created using tf.constant().

```python
import tensorflow as tf

# Scalar
a = tf.constant(10)

# Vector
b = tf.constant([1, 2, 3])

# Matrix
c = tf.constant([[1, 2], [3, 4]])

print(a)
print(b)
print(c)
```
