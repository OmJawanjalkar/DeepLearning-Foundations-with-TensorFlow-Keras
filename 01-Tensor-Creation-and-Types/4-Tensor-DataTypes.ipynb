### Tensor-DataTypes





Every tensor in TensorFlow has a data type (dtype) that defines what kind of values the tensor stores.

```python
import tensorflow as tf

x = tf.constant([1, 2, 3])

print(x.dtype)
```

```python
x = tf.constant([1.0, 2.0, 3.0])

print(x.dtype)
```

### Specify Data Type

```python
x = tf.constant([1, 2, 3], dtype=tf.float32)

print(x)
```

### Change Data Type using tf.cast()

```python
x = tf.constant([1, 2, 3])

y = tf.cast(x, dtype=tf.float32)

print(y)
```

### Common TensorFlow Data Types: tf.int32, tf.int64, tf.float16, tf.float32, tf.float64, tf.bool, and tf.string.

### Key Point: Data types are important because they affect memory usage, computation speed, and model training. In deep learning, float32 is the most commonly used data type.
