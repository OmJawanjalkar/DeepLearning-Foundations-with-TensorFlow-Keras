## Tensor Creation in TensorFlow

A tensor is the basic data structure used in TensorFlow. It is similar to a NumPy array and can store numerical data in different dimensions.

## Tensors can be created using:

      tf.constant() — creates a fixed tensor.
      tf.Variable() — creates a tensor whose values can be changed.
      tf.zeros() — creates a tensor filled with zeros.
      tf.ones() — creates a tensor filled with ones.
      tf.range() — creates a sequence of values.
      tf.random — creates tensors with random values.

```python
import tensorflow as tf
```

```python
# 1. tf.constant() - Fixed tensor
a = tf.constant([1, 2, 3])
print(a)
```

```python
# 2. tf.Variable() - Changeable tensor
b = tf.Variable([10, 20, 30])
b.assign([40, 50, 60])
print(b)
```

```python
# 3. tf.zeros() - Tensor filled with zeros
c = tf.zeros(shape=(2, 3))
print(c)
```

```python
# 4. tf.ones() - Tensor filled with ones
d = tf.ones(shape=(2, 3))
print(d)
```

```python
# 5. tf.range() - Sequence of values
e = tf.range(start=0, limit=10, delta=2)
print(e)
```

```python
# 6. tf.random - Random tensor
f = tf.random.uniform(shape=(2, 3))
print(f)
```

Remember: tf.constant() → fixed, tf.Variable() → changeable, tf.zeros() / tf.ones() → initialized tensors, tf.range() → sequence, tf.random → random values.

##### Key idea: A tensor can have different dimensions: Scalar (0D) → Vector (1D) → Matrix (2D) → Higher-dimensional Tensor (3D+)
