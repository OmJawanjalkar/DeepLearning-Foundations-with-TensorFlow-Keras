### TensorFlow-Variables

  A TensorFlow Variable is a tensor whose value can be changed during program execution. Variables are mainly used to store weights and biases in neural networks.

```python
import tensorflow as tf
```

```python
x = tf.Variable(10)
```

```python
x
```

```python
x.assign(20)
print(x)
```

```python
x.assign_add(1, 1, 1)
print(x)

x.assign_sub(1, 1, 1)
print(x)
```
