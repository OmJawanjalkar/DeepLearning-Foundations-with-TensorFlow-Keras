### What is Gradient Descent?

Gradient Descent is an optimization algorithm used to minimize the loss function by repeatedly updating model parameters (weights).

The basic idea is:

            Calculate Loss → Calculate Gradient → Update Weight → Repeat

Weight Update Formula

       wnew​=wold​−learning_rate × gradient
	​

 −learning_rate×gradient

### Simple TensorFlow Example

```python
import tensorflow as tf
```

```python
w = tf.Variable(5.0)

learning_rate = 0.1
```

```python
for step in range(10):
    with tf.GradientTape() as tape:
        loss = w**2
```

```python
gradient = tape.gradient(loss, w)

w.assign_sub(learning_rate * gradient)

print(f"Step: {step}, Weight: {w.numpy()}, Loss: {loss.numpy()}")
```

### Here, TensorFlow calculates the gradient using GradientTape, and assign_sub() updates the weight toward the minimum loss.

### Key Point: GradientTape calculates gradients, while Gradient Descent uses those gradients to update weights and minimize loss.
