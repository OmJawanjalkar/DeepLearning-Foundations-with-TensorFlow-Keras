## What is GradientTape?

      tf.GradientTape() is a TensorFlow tool used for automatic differentiation. It records mathematical operations and automatically calculates gradients (derivatives).

      Gradients tell us how much the output changes when the input changes.

###      Simple Example

```python
import tensorflow as tf
```

```python
x  = tf.Variable(3.0)

with tf.GradientTape() as tape:
    y= x**2
```

```python
gradient = tape.gradient(y, x)
```

```python
print("The gradient of y with respect to x is:", gradient.numpy())
```

### Because:

      y = x² → dy/dx = 2x → 2(3) = 6

### Why is GradientTape important?

    During neural network training:

    Forward Pass → Calculate Loss → GradientTape calculates gradients → Optimizer updates weights

    Key Point: GradientTape automatically calculates gradients needed for backpropagation and neural network training.
